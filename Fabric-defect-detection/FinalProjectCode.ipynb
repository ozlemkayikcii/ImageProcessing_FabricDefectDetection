```python
import os
import cv2
import numpy as np
from sklearn.svm import OneClassSVM
from sklearn.model_selection import train_test_split

# Google Drive bağlantısını kurmak için gerekli kod
from google.colab import drive
drive.mount('/content/drive')

# Veri klasörü yolu
data_path = '/content/drive/MyDrive/images/'

# Normal verilerin bulunduğu klasör
normal_path = os.path.join(data_path, 'defect_free')

# Anormal verilerin bulunduğu klasör
anomaly_path = os.path.join(data_path, 'stain')

# Verileri depolamak için listeler
normal_images = []
anomaly_images = []

# Normal verileri yükleme
for filename in os.listdir(normal_path):
    img = cv2.imread(os.path.join(normal_path, filename))
    if img is not None:
        img = cv2.resize(img, (1920, 1080))  # Resmi istenen boyuta yeniden boyutlandırma
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Gri tonlamaya dönüştürme
        normal_images.append(img.flatten())  # Düzleştirme ve normal veri listesine ekleme

# Anormal verileri yükleme
for filename in os.listdir(anomaly_path):
    img = cv2.imread(os.path.join(anomaly_path, filename))
    if img is not None:
        img = cv2.resize(img, (1920, 1080))  # Resmi istenen boyuta yeniden boyutlandırma
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Gri tonlamaya dönüştürme
        anomaly_images.append(img.flatten())  # Düzleştirme ve anormal veri listesine ekleme

# Verileri NumPy dizilerine dönüştürme
normal_images = np.array(normal_images)
anomaly_images = np.array(anomaly_images)

# One-class SVM modelini oluşturma
model = OneClassSVM(gamma='scale')

# Sadece normal verileri kullanarak modeli eğitme
model.fit(normal_images)

# Test verilerini oluşturma
test_images = np.concatenate((normal_images, anomaly_images), axis=0)
test_labels = np.concatenate((np.ones(len(normal_images)), -np.ones(len(anomaly_images))), axis=0)

# Test verilerini karıştırma
random_state = np.random.get_state()
np.random.shuffle(test_images)
np.random.set_state(random_state)
np.random.shuffle(test_labels)

# Test verilerini tahmin etme
predictions = model.predict(test_images)

# Anomalileri tespit etme
anomaly_indices = np.where(predictions == -1)[0]
anomaly_images = test_images[anomaly_indices]

# Anomalilerin indekslerini ve tahminlerini yazdırma
print("Anomaly Indices:", anomaly_indices)
print("Predictions:", predictions[anomaly_indices])


```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Metrik ölçümleri hesaplama
accuracy = accuracy_score(test_labels, np.sign(predictions))
precision = precision_score(test_labels, np.sign(predictions))
recall = recall_score(test_labels, np.sign(predictions))
f1 = f1_score(test_labels, np.sign(predictions))

# Metrikleri yazdırma
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
